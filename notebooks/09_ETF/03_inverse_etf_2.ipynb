```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Function for recalculating series with a common base
def rebased(ts):
    return ts / ts[0]
```

```python
# Read data from csv
df = pd.read_csv("../../data/short_etfs.csv", index_col="Date", parse_dates=["Date"])
df.head(5)
```

```python
# Calculate all series starting from first value.
df = df.apply(rebased)
df.head(5)
```

```python
fig = plt.figure(figsize=(12, 8))
```

```python
# The first subplot, planning for 3 plots high, 1 plot wide, this being the first.
ax = fig.add_subplot(111)
ax.set_title("Short ETF Effect")
ax.plot(
    df["SPY"], linestyle="-", label="SPY Index Tracker", linewidth=3.0, color="black"
)
ax.plot(df["SH"], linestyle="-", label="SH Short ETF", linewidth=3.0, color="grey")
ax.plot(
    df["SDS"],
    linestyle="--",
    label="SDS Double Inverse ETF",
    linewidth=2.0,
    color="grey",
)
ax.legend()
```

Try a different starting point:

```python
def plot_comparison(start, end=None):

    df = (
        pd.read_csv("../../data/short_etfs.csv", index_col="Date", parse_dates=["Date"])
        .loc[start:end]
        .apply(rebased)
    )

    fig = plt.figure(figsize=(12, 8))

    # The first subplot, planning for 3 plots high, 1 plot wide, this being the first.
    ax = fig.add_subplot(111)
    ax.set_title("Short ETF Effect")
    ax.plot(
        df["SPY"],
        linestyle="-",
        label="SPY Index Tracker",
        linewidth=3.0,
        color="black",
    )
    ax.plot(df["SH"], linestyle="-", label="SH Short ETF", linewidth=3.0, color="grey")
    ax.plot(
        df["SDS"],
        linestyle="--",
        label="SDS Double Inverse ETF",
        linewidth=2.0,
        color="grey",
    )
    ax.legend()
```

```python
plot_comparison("2014-01-01")
```

```python
plot_comparison("2008-01-01")
```

```python
plot_comparison("2008-01-01", "2013-01-01")
```
